```python
print(dataset['co2'].head(20))
co2_dt = dataset['co2']
print(co2_dt)
sns.set_style('whitegrid')
sns.histplot(data = co2_dt, bins = 20, color = 'blue', edgecolor = 'black')
plt.title('Distribution of Carbon dioxide')
plt.xlabel('Emission')
plt.ylabel('frequency')
plt.savefig('co2_img')
plt.show()
```

```python
print(dataset['cylinders'].head(10))
cyl_data = dataset['cylinders']
print(cyl_data)
sns.set_style('whitegrid')
sns.histplot(data = cyl_data, bins = 20, color = 'red')
plt.title('Numbers of cylinders in each vehicle')
plt.xlabel('cylinders')
plt.ylabel('frequency')
plt.savefig('num_cylinders')
plt.show()
```

**Intepretation**
The histogram shows the number of cylinders in each vehicle

```python
print(dataset['ghgScore'].head(15))
ghgsc_data = dataset['ghgScore']
print(ghgsc_data)
sns.set_style('whitegrid')
sns.histplot(data = ghgsc_data, bins = 20, color = 'black')
plt.xlabel('Green house gas emission')
plt.ylabel('Frequency')
plt.title('Distribution of green house gas  among vehicles')
plt.savefig('ghg_dist')
plt.show()
```

**Interpretation**
This histogram indicates the emission of green house gas and the frequency of the emission among vehicles

```python
values = dataset.select_dtypes(include= ['number'])
print(values)
corr_matrix = values.corr()
sns.heatmap(data = corr_matrix, annot= True, cmap= 'coolwarm', fmt= '.2f')
plt.title('Correlation map')
plt.figure(figsize=(10,8))
plt.savefig('corr_map')
plt.show()
```

**Interpretation**
The visualization indicates the relationship between different features and greenhouse gas emissions, Features with highly correlation indicate stronger correlations with  much predictive power for greenhouse gas emissions while weakly correlated features may not provide much power of green house emissions
